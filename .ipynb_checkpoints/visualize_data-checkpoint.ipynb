```python
proc_img = img[:, :, 0].copy()
thrs = 60
proc_img[proc_img <= thrs] = 0
proc_img[proc_img > thrs] = 255
plt.figure(figsize=(10, 10))
plt.imshow(proc_img)
```

```python
# convert img to grayscale
proc_img_3 = np.repeat(proc_img[:, :, np.newaxis], 3, axis=2)
gray = cv2.cvtColor(proc_img_3, cv2.COLOR_BGR2GRAY)

# do adaptive threshold on gray image
#thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 101, 3)
#plt.imshow(thresh)
#plt.show()

# apply morphology open then close
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
#blob = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (9,9))
blob = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)
blob = cv2.morphologyEx(blob, cv2.MORPH_OPEN, kernel)

# invert blob
blob = (255 - blob)
plt.figure(figsize=(10, 10))
plt.imshow(blob)
```

```python
plt.figure(figsize=(10, 10))
# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(blob, (3,3), 0)
# Canny Edge Detection
#edges = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)
edges = cv2.Canny(image=img_blur, threshold1=100, threshold2=200) # Canny Edge Detection
# Display Canny Edge Detection Image
plt.imshow(edges)
```

```python
# Get contours
cnts = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# test blob size
blob_area = 5000
print(blob_area)
new_cnts = []
rmv_cnts = []
blob_areas = []
for cont in cnts[1]:
    if (cv2.contourArea(cont) >= blob_area):
        new_cnts.append(cont)
        blob_areas.append(cv2.contourArea(cont))
    else:
        rmv_cnts.append(cont)
print(blob_areas)
print(len(blob_areas))
# draw contour
result = img.copy()
contour = cv2.drawContours(result, new_cnts, -1, (255,255,255), 1)
fill_img = cv2.fillPoly(np.ones_like(img) * 255, pts=new_cnts, color=(0,0,0))
plt.figure(figsize=(10, 10))
plt.imshow(fill_img)
plt.imsave("figs/centriols_mask_ex.png", fill_img)
```
