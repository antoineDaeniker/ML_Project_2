```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cv2
%pylab inline
import matplotlib.image as mpimg
import imageio
```

```python
# get all images
data = pd.read_csv("datasets_full/annotations.csv")
print(np.unique(data['image_name']))
```

```python
# check if for every image there are 3 channels
image_list = np.unique(data['image_name'])

for img_name in image_list:
    for channel in [1,2,3]:
        if (img_name[:-1] + str(channel)) not in image_list:
            print("Missing channel for: ", (img_name[:-1] + str(channel)))
            
```

Channel 1 is missing for RPE1wt_CP110+GTU88+PCNT_2_000_004_max

Create one png and tif from the three channels.

For the missing channel 1 for RPE1wt_CP110+GTU88+PCNT_2_000_004_max leave that empty

```python
# function to give the right path back according to name from the annotation csv 
data_folder = "datasets_full"
data_folders = ["RPE1wt_CEP63+CETN2+PCNT_1", "RPE1wt_CEP152+GTU88+PCNT_1", "RPE1wt_CP110+GTU88+PCNT_2"]
proj_ch = "projections_channel"
channels = [["CEP63", "CETN2", "PCNT"], ["CEP152", "GTU88", "PCNT"], ["CP110", "GTU88", "PCNT"]]
image_type = ["png", "tif"]

def give_path(name, tif=True):
    # remove start and end - as always the same lenght
    ch_num = int(name[-1:])
    #print(ch_num)
    ch = name[:-15]
    folder = ch
    ch = ch[7:]
    ch = ch[:-2]
    #print(ch)
    ch_list = ch.split("+")
    #print(folder)
    #print(ch_list[ch_num - 1])
    
    if tif == True:
        path = data_folder + "/" + folder + "/" + proj_ch + "/" + ch_list[ch_num - 1] + "/" + "tif" + "/" + name
        path += ".tif"
    else:
        path = data_folder + "/" + folder + "/" + proj_ch + "/" + ch_list[ch_num - 1] +  "/" + "png" + "/" + name
        path += ".png"
    return path

# ML_Project_2\datasets_full\RPE1wt_CP110+GTU88+PCNT_2\projections_channel\GTU88\tif\...

example_name = "RPE1wt_CP110+GTU88+PCNT_2_004_004_max_C3"
give_path(example_name)
```

```python
# show an png and tif
img = mpimg.imread(give_path('RPE1wt_CP110+GTU88+PCNT_2_000_004_max_C3', tif=False))
imgplot = plt.imshow(img)
plt.show()
```

CREATE PNG

```python
# create image names without channel names
names = []
for img_name in image_list:
    names.append(img_name[:-3])

img_name_uniq = np.unique(names)
```

```python
# create one png 

png_path = "./datasets_full/all_channel_img/png/"

for img_name_woch in img_name_uniq:
    print("Creating from", img_name_woch)
    img_channels = []
    missing_ch = []
    for ch in ["_C1", "_C2", "_C3"]:
        img_c = img_name_woch + ch
        img_path = give_path(img_c, tif=False)
        if os.path.isfile('./' + img_path):
            im = imageio.imread(img_path)
            img_channels.append(im)
        else:
            # missing channel, fill with 0
            im = np.zeros((2048, 2048))
            img_channels.append(im)
    c_123 = np.dstack((img_channels))
    # save
    name = img_name_woch + ".png"
    imageio.imsave(png_path + name, c_123)
    #plt.imshow(c_123, interpolation='nearest')
    #plt.show()
    #break
```

```python
# create one tif 

png_path = "./datasets_full/all_channel_img/tif/"

for img_name_woch in img_name_uniq:
    print("Creating from", img_name_woch)
    img_channels = []
    missing_ch = []
    for ch in ["_C1", "_C2", "_C3"]:
        img_c = img_name_woch + ch
        img_path = give_path(img_c, tif=True)
        if os.path.isfile('./' + img_path):
            im = imageio.imread(img_path)
            img_channels.append(im)
        else:
            # missing channel, fill with 0
            im = np.zeros((2048, 2048))
            img_channels.append(im)
    c_123 = np.dstack((img_channels))
    # save
    name = img_name_woch + ".tif"
    imageio.imsave(png_path + name, c_123)
    #plt.imshow(c_123, interpolation='nearest')
    #plt.show()
    #break
```
